# The Real Problem

One of the key activities of any IT function is to “Keep the lights on” to ensure there is no impact to the
Business operations. IT leverages Incident Management process to achieve the
above Objective. An incident is something that is unplanned interruption to an IT service or
reduction in the quality of an IT service that affects the Users and the Business. The main goal
of Incident Management process is to provide a quick fix / workarounds or solutions that resolves the
interruption and restores the service to its full capacity to ensure no business impact. In most of the
organizations, incidents are created by various Business and IT Users, End Users/ Vendors if they have
access to ticketing systems, and from the integrated monitoring systems and tools. Assigning the
incidents to the appropriate person or unit in the support team has critical importance to provide
improved user satisfaction while ensuring better allocation of support resources. The assignment of
incidents to appropriate IT groups is still a manual process in many of the IT organizations. Manual
assignment of incidents is time consuming and requires human efforts. There may be mistakes due to
human errors and resource consumption is carried out ineffectively because of
the misaddressing. On the other hand, manual assignment increases the response and resolution times
which result in user satisfaction deterioration / poor customer service.

# Business Domain Value

In the support process, incoming incidents are analyzed and assessed by organization’s support teams to
fulfill the request. In many organizations, better allocation and effective usage of the valuable support
resources will directly result in substantial cost savings.


Currently the incidents are created by various stakeholders (Business Users, IT Users and Monitoring
Tools) within IT Service Management Tool and are assigned to Service Desk teams (L1 / L2 teams). This
team will review the incidents for right ticket categorization, priorities and then carry out initial
diagnosis to see if they can resolve. Around ~54% of the incidents are resolved by L1 / L2 teams. Incase
L1 / L2 is unable to resolve, they will then escalate / assign the tickets to Functional teams from
Applications and Infrastructure (L3 teams). Some portions of incidents are directly assigned to L3 teams
by either Monitoring tools or Callers / Requestors. L3 teams will carry out detailed diagnosis and resolve
the incidents. Around ~56% of incidents are resolved by Functional / L3 teams. Incase if vendor support
is needed, they will reach out for their support towards incident closure.


L1 / L2 needs to spend time reviewing Standard Operating Procedures (SOPs) before assigning to
Functional teams (Minimum ~25-30% of incidents needs to be reviewed for SOPs before ticket
assignment). 15 min is being spent for SOP review for each incident. Minimum of ~1 FTE effort needed
only for incident assignment to L3 teams.

During the process of incident assignments by L1 / L2 teams to functional groups, there were multiple
instances of incidents getting assigned to wrong functional groups. Around ~25% of Incidents are
wrongly assigned to functional teams. Additional effort needed for Functional teams to re-assign to right
functional groups. During this process, some of the incidents are in queue and not addressed timely
resulting in poor customer service.
Guided by powerful AI techniques that can classify incidents to right functional groups can help
organizations to reduce the resolving time of the issue and can focus on more productive tasks.

```python
from nltk.corpus import stopwords
stoplist = set(stopwords.words("english"))

def preprocess_text(df, column_name=''):
    
    # Replace empty strings with Null
    df[column_name].replace('', np.nan, inplace = True)

    # Drop Null values
    df = df.dropna()
    
    # Remove the numbers
    df[column_name] = df[column_name].str.replace('\d+', '')

    # Select only alphabets
    df[column_name] = df[column_name].apply(lambda x: re.sub('[^A-Za-z]+', ' ', x))

    # Convert text to lowercase
    df[column_name] = df[column_name].apply(lambda x: x.lower())

    # Strip unwanted spaces
    df[column_name] = df[column_name].apply(lambda x: x.strip())

    # Remove stopwords
    df[column_name] = df[column_name].apply(lambda x: ' '.join([word for word in x.split() if word not in stoplist]))
    
#     #remove x000d from text.
#     df[column_name] = df[column_name].str.replace('_x000D_',' ')
        
#     # remove ? from text.
#     df[column_name] = df[column_name].str.replace('?',' ')
    
#     # correct acct to account.
#     df[column_name] = df[column_name].str.replace('acct', 'account')
    
#     # correct fw: to forward.
#     df[column_name] = df[column_name].str.replace('fw:', 'forward')
    
#     # correct fw to forward.
#     df[column_name] = df[column_name].str.replace('fw','forward')    
    
#     # Remove hex chars
#     df[column_name] = df[column_name].str.replace('==pcap 1 hex s==[\w\W]*==pcap 1 hex e==', ' ')
    
#     # Remove ascii chars
#     df[column_name] = df[column_name].str.replace('==pcap 1 ascii s==[\w\W]*==pcap 1 ascii e==', ' ')
    
#     # Remove correlation data
#     df[column_name] = df[column_name].str.replace('\[correlation_data\](.*\n){2}', ' ')
    
#     # Remove correlation data
#     df[column_name] = df[column_name].str.replace('\[correlation_data\]', ' ')
    
#     # Remove [no entry] 
#     df[column_name] = df[column_name].str.replace('\[no entry\]', ' ')
    
#     # Remove ascii packet(s):
#     df[column_name] = df[column_name].str.replace('ascii packet(s):', ' ')
    
#     # Remove hex packet(s)::
#     df[column_name] = df[column_name].str.replace('hex packet(s):', ' ')
    
#     # remove tabs
#     df[column_name] = df[column_name].str.replace('\t', '')    
#     df[column_name] = df[column_name].str.replace('\r', '')
        
#     # Remove new lines.
#     df[column_name] = df[column_name].str.replace('\n', '')
    
#     # We will use google translate to translate the sentence to english.
#     if(len(df[column_name].tolist()) > 4000):
#         n = 3000
#         GtransStrings = []
#         for index in range(0, len(df[column_name].tolist()), n):
#             separated_strings = df[column_name].tolist()[index : index + n]
            
#             GtransStrings.append(Gtrans.translate(separated_strings[0]))
#         df[column_name] = df[column_name].str=" ".join(GtransStrings)
        
#     else:
#         if(df[column_name].str!=''):
#             df[column_name]=Gtrans.translate(df[column_name].str)

    return df

# Lemmatization implemented
lemmatizer = WordNetLemmatizer()
wordnet_lemmatizer = WordNetLemmatizer()

# Convert NLTK to word tags
def nltkToWordTag(inpNltkTag):
    if inpNltkTag.startswith('J'):
        return wordnet.ADJ
    elif inpNltkTag.startswith('V'):
        return wordnet.VERB
    elif inpNltkTag.startswith('N'):
        return wordnet.NOUN
    elif inpNltkTag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

def lemmatize_sentence(inpSentence):
    # Tokenize the sentence and find the POS tag for each token
    nltk_Tag = nltk.pos_tag(nltk.word_tokenize(inpSentence))  
    wordnet_Tag = map(lambda x: (x[0], nltkToWordTag(x[1])), nltk_Tag)
    lemmatized_sentence = []
    for word, tag in wordnet_Tag:
        if tag is None:
            # If there is no available tag, append the token as is
            lemmatized_sentence.append(word)
        else:        
            # Else apply the appropriate tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

```python
# Delete related words
email_word_list = ['.com','thanks','gmail','cc','hi','hello','regards']

def removeMailWords(inpText):
    finalTokens = []
    tokens = nltk.word_tokenize(inpText)
    for w in tokens:
        if (w not in email_word_list):
            finalTokens.append(w)
    inpText = " ".join(finalTokens)
    return inpText
```

```python
def train_model(classifier, featureVectorTrain, label, featureVectorValid, isNeuralNetwork=False):
    # Fit the training dataset on the classifier
    classifier.fit(featureVectorTrain, label)
    
    # Predict the labels on validation dataset
    predictions = classifier.predict(featureVectorValid)
    
    if isNeuralNetwork:
        predictions = predictions.argmax(axis=-1)
    
    return classifier, predictions

def cal_accuracy(model_name, y_test, y_pred): 

    print ("############  Model Used: ", model_name, "####################")
    print("Confusion Matrix:\n ", metrics.confusion_matrix(y_test, y_pred))       
    print ("Accuracy : ", metrics.accuracy_score(y_test,y_pred)*100) 
    
    print("Recall: {:.2f}".format(metrics.recall_score(y_test, y_pred)))
    print("Precision: {:.2f}".format(metrics.precision_score(y_test, y_pred)))      
    print("Report : ", metrics.classification_report(y_test, y_pred))
    
# Fit & print the accuracy stats for the model passed as an argument
def SummarizeModelOutput(X_train, y_train, X_test, y_test, model, modelType):
    y_pred = model.predict(X_test)
    prediction_train = model.predict(X_train)
    prediction_test = model.predict(X_test)

    dict_results = {"Algorithm": modelType,
                         "Train Accuracy Score": metrics.accuracy_score(y_train, prediction_train.round()) * 100,
                         "Test Accuracy Score": metrics.accuracy_score(y_test, prediction_test.round()) * 100,
                         "Train Recall Score":(metrics.recall_score(y_train, prediction_train.round())),
                         "Test Recall Score":(metrics.recall_score(y_test, prediction_test.round())),
                         "Train Precision":(metrics.precision_score(y_train, prediction_train.round())),
                         "Test Precision":(metrics.precision_score(y_test, prediction_test.round()))                 
                       }

    return pd.Series(dict_results).to_frame().T
```

```python
# In an “one-to-rest” model, we could build multiple independent classifiers and, for an unseen instance, 
# choose the class for which the confidence is maximized

# A suitable or preferred approach to solving multi-label problem is to 
# Decompose it into multiple independent binary classification problems (one per category).

def buildAndTrainModel(X_train, y_train, X_valid = None, y_valid = None, C = 1.0, model = 'lr'):
    if model == "lr":
        model = LogisticRegression(multi_class='multinomial', solver='newton-cg')
#         model = OneVsRestClassifier(model)
        model.fit(X_train, y_train)
    elif model == "svm":
        model = LinearSVC(C=C, penalty='l1', dual=False, loss='squared_hinge')
        #model = OneVsRestClassifier(model)
        model.fit(X_train, y_train)
    elif model == "nbayes":
        model = MultinomialNB(alpha=1.0)
        # model = OneVsRestClassifier(model)
        model.fit(X_train, y_train)
    elif model == "random":
        model = RandomForestClassifier()
        # model = OneVsRestClassifier(model)
        model.fit(X_train, y_train)
    elif model == "gradient":
        params = {'n_estimators': 10, # change to 9000 to obtain 0.505 on LB (longer run time expected)
            'max_depth': 20,
            'min_samples_split': 5,
            'min_samples_leaf': 5,
            'learning_rate': 0.01,
            'max_features':  'sqrt',
            'subsample': 0.8,
            'criterion': 'friedman_mse',
            'loss': 'deviance',
            'random_state':7,
            'verbose':True}
        model = GradientBoostingClassifier(**params)
        # model = OneVsRestClassifier(model)
        model.fit(X_train, y_train)
    return model
```

```python
from nltk.corpus import stopwords

# Define function for cleaning.
def dataCleaner(text):
    
    #remove x000d from text.
    text=text.replace('_x000D_',' ')
    
    #remove ? from text.
    text=text.replace('?',' ')
    
    #correct acct to account.
    text=text.replace('acct','account')
    
    #correct fw to forward.
    text=text.replace('fw:','forward')
    
    #correct fw to forward.
    text=text.replace(' fw ',' forward ')
    
    #Remove hex chars
    lst = re.findall('==pcap 1 hex s==[\w\W]*==pcap 1 hex e==', text)
    for i in lst:
        text=text.replace(i,' ')
        
    #Remove ascii chars
    lst = re.findall('==pcap 1 ascii s==[\w\W]*==pcap 1 ascii e==', text)
    for i in lst:
        text=text.replace(i,' ')
        
    #Remove correlation data 
    lst = re.findall('\[correlation_data\](.*\n){2}', text)
    for i in lst:
        text=text.replace(i,' ')
       
    #Remove [correlation_data]
    lst = re.findall('\[correlation_data\]', text)
    for i in lst:
        text=text.replace(i,' ')
        
    #Remove [no entry] 
    lst = re.findall('\[no entry\]', text)
    for i in lst:
        text=text.replace(i,' ')
        
    #Remove ascii packet(s):
    text=text.replace('ascii packet(s):',' ')
    
    #Remove hex packet(s)::
    text=text.replace('hex packet(s):',' ')
        
    # Removing white spaces and trimmign the space around words.
    text = re.sub(' +', ' ', text.strip())
    
    #removee tabs
    text=text.replace('\t',' ')
    
    text=text.replace('\r',' ')
    
    #Remove new lines.
    text=text.replace('\n',' ')
    
    # We will use google translate to translate the sentence to english.
    if(len(text)>4000):
        n=3000
        GtransStrings = []
        for index in range(0, len(text), n):
            separated_strings = text[index : index + n]
            GtransStrings.append(Gtrans.translate(separated_strings))
        text=" ".join(GtransStrings)
        
    else:
        if(text!=''):
            text=Gtrans.translate(text)
            
    # Changing to lower case all the words.
    text = text.lower()
    text = text.split()
    
    # Removing english stop words.
    text = [word.strip() for word in text if not word in set(stopwords.words('english'))]
    
    # lemmatizing the words to their base forms
    text = [lemma.lemmatize(word) for word in text]
    
    # Joining back the text.
    text = ' '.join(text)
   
    return text
```

```python
# Generate word embeddings
# Construct the embedding weight matrix using the word embeddings via the tokenizer
embeddings_index = dict()
BATCH_SIZE = 1024
EPOCHS = 15

f = open('./glove.6B.300d.txt', encoding="utf8")
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

embedding_matrix = np.zeros((vocab_size, EMBEDDING_DIM))
for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
            
print('Found %s word vectors.' %len(embeddings_index))
```

```python
# Composite model prediction implementation
def getGroupAssignment(text):
    text=dataCleaner(text)
    if text.strip() == '' :
        return 'GRP_0'
    X1=pad_sequences(tokenizer.texts_to_sequences([text]), maxlen = max_len)
    model1Pred=FullEncoder.inverse_transform(np.round(model1.predict(X1),0).astype(int))
    
    if model1Pred=='GRP_0':
        return 'GRP_0'
    else:
        X2=vectorizer.transform([text])
        X2=X2.toarray()
        try:
            model2Pred=smallEncoder.inverse_transform(model2.predict(X2))
            model2Pred=model2Pred[0]
        except:
            model2Pred = 'GRP_0'
        return model2Pred
    
    return 'GRP_0'
```

```python
# Import the ktrain library to train the BERT model
import ktrain
from ktrain import text
```

```python
# Read the dataset & the semi cleaned file
ticket = pd.read_excel('input_data automated ticket dataset.xlsx')
ticket['Description'].fillna(ticket['Short description'],inplace=True)

# Read the cleaned dataset
ticket = pd.read_csv('./semicleaned2.csv')

ticket.drop(['Caller','Description','Short description'],axis=1,inplace=True)
ticket.dropna(inplace=True)
ticket['Group_Mod'] = ticket['Assignment group'].apply(lambda x: 'GRP_X' if x != 'GRP_0' else x)
ticket.head()
```

```python
# Split into X & y
XFull = ticket['CleanedDescription']
yFull = ticket['Group_Mod']
print(XFull.shape)
print(yFull.shape)
```

```python
# Split into train & test with 70% & 30% of data composition respectively
X_train, X_test, y_train, y_test = train_test_split(XFull, yFull, test_size = 0.30)
```

```python
# Convert the train & test sets into arrays
x_train=np.array(X_train)
y_train=np.array(y_train)
x_test=np.array(X_test)
y_test=np.array(y_test)
```

```python
# Convert the text data directly into features for the model that is needed
# Pre-processing logic need not be performed manually as they will be handled by the library implicitly 
# Automatically downloads the pre-trained BERT model along with its vocabulary
# Need to specify the pre-process mode as BERT to process the text in a specific way
(X_train_bert, y_train_bert), (X_test_bert, y_test_bert), preproc = text.texts_from_array(x_train, y_train, x_test, y_test, preprocess_mode='bert', lang='en', maxlen=75)
```

```python
# Loads the pre-trained model with a randomly initialized final Dense layer
# Dense layer will get updated during the training process
# we have not frozen any layers and all the layers of the model are trainable, 
# the weights of all the layers of the model will be updated during backpropagation
model = text.text_classifier('bert', train_data=(X_train_bert, y_train_bert), preproc=preproc)
learner = ktrain.get_learner(model, train_data=(X_train_bert, y_train_bert), val_data=(X_test_bert, y_test_bert), batch_size=64)
```

```python
# For training the model, we will 1st find the optimal learning rate that suits well to our problem statement
# ktrain offers a very good method namely "lr_find" that trains the model at 
# different learning rates & plots loss of the model as the learning rate is increased
# Takes a long time to train usually
learner.lr_find(max_epochs=1, start_lr=1e-5, show_plot=True)
```

```python
# Train the model using the autofit method 
# Classifier is trained & automatically selects the best performing model preventing the underfit & overfit of the model
# We observe the model performance is improving & turning out to be one of the best trained model
# Commenting out the autofit as it takes a very long time to complete its training
# learner.autofit(1e-5)

# Train using 1cycle learning rate schedule for 2 epochs itself yielding an accuracy value of more than 80% 
learner.fit_onecycle(2e-5, 2)
```

```python
# Extracts the predictor variable
# Pass the model & preproc object in the get_predictor method
# The predictor variable can then be used to make predictions on the test data
predictor = ktrain.get_predictor(learner.model, preproc)
```
